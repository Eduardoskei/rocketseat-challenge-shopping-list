```python
items = []

def main():
    while True:
        print('\nMenu:')
        print('1. Adicionar item')
        print('2. Listar todos os itens')
        print('3. Marcar item como comprado')
        print('4. Remover item')
        print('5. Editar item')
        print('6. Lista itens não comprados')
        print('7. Sair')

        action = input('Escolha a ação desejada: ')

        if action == '1':
          try:
            name = input('Digite o nome do item: ')
            quantity = int(input('Digite a quantidade de itens: '))
            add_item(name, quantity)
          except ValueError:
            print('\nEntrada inválida! Por favor, digite um número inteiro')
        elif action == '2':
          get_items()
        elif action == '3':
          try:
            get_items()
            index = int(input('Digite o índice do item que será marcado: '))
            check_item(index - 1)
          except ValueError:
            print('\nEntrada inválida! Por favor, digite um número inteiro')
        elif action == '4':
          try:
            get_items()
            index = int(input('Digite o índice do item que será removido: '))
            remove_item(index - 1)
          except ValueError:
            print('\nEntrada inválida! Por favor, digite um número inteiro')
        elif action == '5':
          try:
            get_items()
            index = int(input('Digite o índice do item que será editado: '))
            edit_item(index - 1)
          except ValueError:
            print('\nEntrada inválida! Por favor, digite um número inteiro')
        elif action == '6':
          get_dont_buy_items()
        elif action == '7':
          print('Saindo do sistema...')
          break
        else:
          print('\nOpção inválida! Por favor, tente novamente.')

def add_item(name, quantity):
  if not name:
    print('\nNome não pode ser vazio. Tente novamente')
    return

  items.append({
      'name': name,
      'quantity': quantity,
      'is_buy': False
  })
  print(f'\nItem "{name}" adicionado com sucesso.')

def get_items():
  if not items:
    print('\nNenhum item cadastrado.')
    return

  print('\nLista de itens não comprados:')
  for i, item in enumerate(items):
    stats = '[X]' if item['is_buy'] == True else '[ ]'
    print(f'{i + 1}. {stats} {item['name']} - {item['quantity']}')

def get_dont_buy_items():
  if not items:
    print('\nNenhum item cadastrado.')
    return

  print('\nLista de itens:')
  for i, item in enumerate(items):
    stats = '[X]' if item['is_buy'] else '[ ]'

    if stats == '[ ]':
      print(f'{i + 1}. {stats} {item['name']} - {item['quantity']}')

def edit_item(index):
  if index < 0 or index > len(items):
    print('\nÍndice inválido! Tente novamente.')
    return

  try:
    new_quantity = int(input('Digite a nova quantidade: '))
  except ValueError:
    print('\nEntrada inválida! Por favor, digite um número inteiro')
    return

  items[index]['quantity'] = new_quantity
  print(f'\nItem "{items[index]['name']}" foi editado com sucesso.')
def check_item(index):
  if index < 0 or index > len(items):
    print('\nÍndice inválido! Tente novamente.')
    return

  items[index]['is_buy'] = True
  print(f'\nItem "{items[index]['name']}" foi marcado como comprado')

def remove_item(index):
  if index < 0 or index > len(items):
    print('\nÍndice inválido! Tente novamente.')
    return

  name = items.pop(index)['name']
  print(f'\nItem "{name}" foi removido com sucesso.')
```

```python
main()
```
